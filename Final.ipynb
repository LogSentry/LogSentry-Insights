```python
import numpy as np 

class Perseptron: 
    def __init__(self, number_of_inputs, lr = 0.01):
        self.weights = np.random.rand(number_of_inputs) # set seights to a random number between 0-1 for number of inputs
        self.bais = 1 # set bais to one, only one bais in a perseptron
        self.learning_rate = lr

    # predie the values
    def forward(self, inputs):
        weighted_sum = np.dot(inputs, self.weights)  + self.bais  # x1 * w1 + x2 * w2 ... + b
        return self.sigmoid(weighted_sum) # retrun activated funtion
    
    #claculate error and fix wights and bias based on error 
    def backward(self, predidtion, ground_truth, input): 
        error = ground_truth - predidtion # since it is ground_truth - predidtion and not predidtion - ground_truth, we add to adjust the weight and bias, not subtract
        self.weights += error * self.learning_rate * input # The gradient of the loss function with respect to each weight is proportional to the input feature
        self.bais += error * self.learning_rate # update bais with the error 
    
    def train(self, inputs, ground_truths):
        for input, ground_truth in zip(inputs,ground_truths): # iterate over both inputs and ground truths simultaneously 
            prediction = self.forward(input)
            self.backward(prediction, ground_truth, input)


    #helper funtions 
    def sigmoid(self, weighted_sum):
        return 1 / (1 + np.exp(-weighted_sum))
```
